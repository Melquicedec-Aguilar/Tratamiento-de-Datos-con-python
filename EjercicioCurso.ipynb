```python
import pandas as pd
import numpy as np
```

```python
df_titanic = pd.read_csv("https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv")
df_titanic.info()
```

```python
df_titanic
```

```python
#Se visualiza el numero de datos faltantes (NaN)
df_titanic.isnull().sum()
```

####Si la edad es null (na) llenar el campo con la media de la edad (usar el método fillna())

```python
df_titanic['age'].describe()
```

```python
df_titanic['age'].fillna((df_titanic['age'].mean()), inplace= True)
```

####Si embarqued es na llenar con **S**

```python
df_titanic['embarked'].fillna(value= 'S', inplace= True)
```

```python
df_titanic.isnull().sum()
```

####Cambiar los valores alfanumericos de sex a numero usando { 'female': 1, 'male': 0 }

```python
#Reemplazamos los valores de la columna sex
df_titanic['sex'] = df_titanic['sex'].replace({'female':1, 'male':0})
```

####Cambiar los valores alfanumericos de embarked a numero usando { 'S': 0, 'C': 1, 'Q': 2 }

```python
#Reemplazamos los valores de la columna embarked
df_titanic['embarked'] = df_titanic['embarked'].replace({'S': 0, 'C': 1, 'Q': 2 })
```

```python
df_titanic
```

####Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas

```python
df_titanic.loc[df_titanic['age'] >= 16, 'child'] = 0
df_titanic.loc[df_titanic['age'] < 16, 'child'] = 1
df_titanic
```

####Eliminar las columnas ticket, cabin y sibsp

```python
df_titanic1 = df_titanic.drop(['ticket', 'cabin', 'sibsp'], axis=1)
```

```python
df_titanic1
```

####Validar que no existan registros duplicados

```python
#N hay registros duplicados
df_dup = df_titanic.duplicated()
df_titanic[df_dup]
```

####Listar los valores únicos del campo sex

```python
df_titanic['sex'].unique()
```

####Quitar los espacios vacíos de los campos que no sean numéricos

```python
df_titanic1.info()
```

```python
df_titanic1['name'] = df_titanic1['name'].str.strip()
df_titanic1['boat'] = df_titanic1['boat'].str.strip()
df_titanic1['home.dest'] = df_titanic1['home.dest'].str.strip()
df_titanic1
```

####Cambiar los nombres de las variables a español sin espacios

```python
df_titanic1.columns
```

```python
df_titanic1.columns = ['TipoDeBoleto', 'Sobreviviente', 'Nombre', 'Genero', 'edad', 'NumeroDePadres', 'CostoBoleto', 'Puerto',
                      '#Botes', '#Cuerpos', 'OrigenYDestino', 'niño']
```

####Guardar los datos limpios en un archivos json orientado a registros llamado iniciales_titanic_clean.json (sus iniciales)

```python
df_titanic1.to_json('MAG_titanic_clean.json', orient='records')
```

####Guardar los datos limpios en un archivo de excel con dos hojas una con los sobrevientes y otra con los fallecidos llamado iniciales_titanic_clean.xlsx (sus iniciales)

```python
df_sobrevivientes = df_titanic1.loc[df_titanic1['Sobreviviente'] == 1]
```

```python
df_fallecidos = df_titanic1.loc[df_titanic1['Sobreviviente'] == 0]
```

```python
with pd.ExcelWriter('MAG_titanic_clean.xlsx') as w:
  df_sobrevivientes.to_excel(w, sheet_name='Sobrevivientes', index=False)
  df_fallecidos.to_excel(w, sheet_name='Fallecido', index=False)
```
